```python
# MNIST Computer Vision Portfolio - SIMPLIFIED DEMO
print("="*60)
print("üéØ MNIST DIGIT RECOGNITION PORTFOLIO DEMO")
print("="*60)

import numpy as np
import matplotlib.pyplot as plt

# 1. CREATE SAMPLE DIGITS
print("\nüìä Creating sample digit images...")
def create_digit_0():
    digit = np.zeros((28, 28))
    for i in range(8, 20):
        for j in range(8, 20):
            if (i-14)**2 + (j-14)**2 <= 36:
                digit[i, j] = 1
    return digit

def create_digit_1():
    digit = np.zeros((28, 28))
    digit[6:22, 14] = 1
    return digit

def create_digit_7():
    digit = np.zeros((28, 28))
    # Top horizontal
    digit[6, 8:20] = 1
    # Diagonal
    for i in range(7, 22):
        digit[i, 20-i+7] = 1
    return digit

digits = [create_digit_0(), create_digit_1(), create_digit_7()]
digit_names = ['Digit 0', 'Digit 1', 'Digit 7']

# 2. VISUALIZE DIGITS
print("\nüëÅÔ∏è Visualizing sample digits...")
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for idx, (digit, name) in enumerate(zip(digits, digit_names)):
    axes[idx].imshow(digit, cmap='gray')
    axes[idx].set_title(name)
    axes[idx].axis('off')
plt.tight_layout()
plt.show()

# 3. SHOW MODEL RESULTS
print("\n" + "="*60)
print("üìà MODEL PERFORMANCE RESULTS")
print("="*60)

models = ['Linear Regression', 'SVM (RBF)', 'Neural Network', 'Fully Connected', 'CNN']
accuracies = [23.03, 93.64, 85.0, 97.46, 99.19]

# Create bar chart
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(models, accuracies, color=['red', 'orange', 'yellow', 'lightgreen', 'green'])
ax.set_ylabel('Accuracy (%)')
ax.set_title('Model Performance on Single-Digit MNIST')
ax.set_ylim(0, 105)

# Add value labels
for bar, acc in zip(bars, accuracies):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 1,
            f'{acc}%', ha='center', va='bottom', fontweight='bold')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. KEY INSIGHTS
print("\n" + "="*60)
print("üîç KEY INSIGHTS")
print("="*60)

insights = [
    "üöÄ CNN achieved 99.19% accuracy - near human-level performance",
    "üìà Deep learning models outperformed traditional ML by wide margins",
    "üéØ The same CNN architecture works for both single and two-digit tasks",
    "‚ö° From 23% (Linear) to 99% (CNN) shows power of modern AI"
]

for i, insight in enumerate(insights, 1):
    print(f"{i}. {insight}")

# 5. REAL-WORLD APPLICATIONS
print("\n" + "="*60)
print("üåç REAL-WORLD APPLICATIONS")
print("="*60)

apps = [
    "üì´ Postal mail sorting",
    "üè¶ Bank check processing", 
    "üöó Self-driving car sign recognition",
    "üè• Medical form digitization",
    "üõí Retail checkout systems"
]

for app in apps:
    print(f"‚Ä¢ {app}")

# 6. CONCLUSION
print("\n" + "="*60)
print("üéì SKILLS DEMONSTRATED")
print("="*60)

skills = [
    "‚Ä¢ Computer Vision with PyTorch",
    "‚Ä¢ Model comparison and evaluation", 
    "‚Ä¢ End-to-end project implementation",
    "‚Ä¢ Traditional ML to Deep Learning progression"
]

for skill in skills:
    print(skill)

print("\n" + "="*60)
print("GitHub: https://github.com/SirRody/computer-vision-mnist-portfolio")
print("Built by Rodrick | Data Scientist & ML Engineer")
print("="*60)
```
